```python
def step_gradient(X, Y, m, learning_rate):
    m_slope = np.zeros(len(X[0]))
    M = len(X)
    for i in range(M):
        x = X[i]
        y = Y[i]
        for j in range(len(x)):
            m_slope[j] += -(2 / M) * (y - sum(m * x)) * x[j]
    new_m = m - (learning_rate * m_slope)
    return new_m

def cost(m, x, y):
    cost = 0
    M = len(x)
    for i in range(M):
        cost += (1 / M) * ((y[i] - sum(m * x[i])) **2)
    print(cost)

def gd(x, y, learning_rate, iterations):
    m=np.zeros(len(x[0]))
    for i in range(iterations):
        m = step_gradient(x, y, m, learning_rate)
        print("itr= ", i, "cost=", end=' ')
        cost(m, x, y)
    return m

def run(x, y):
    iterations = 40 #50
    learning_rate = 0.35
    x=np.append(x, np.ones(len(x)).reshape(-1, 1), axis=1)
    m=gd(x, y, learning_rate, iterations)
    return m
```

```python
X = train_data[:, :-1]
print(X.shape)
Y = train_data[:, -1]
print(Y.shape)
```

```python
X
```

```python
from sklearn import preprocessing
from sklearn import model_selection

scaler = preprocessing.StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
X
```

```python
m = run(X, Y)
```

```python
X_test = test_data
X_test = scaler.transform(X_test)

X_test = np.append(X_test, np.ones(len(X_test)).reshape(-1, 1), axis = 1)

Y_pred = []
for i in X_test:
    Y_pred.append(sum(m * i))

for i in Y_pred:
    print(i)

Y_pred = np.array(Y_pred)

np.savetxt(X = Y_pred, fname = "Predicted.csv", delimiter = ",", fmt = "%.5f")
```
